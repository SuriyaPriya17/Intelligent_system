```python
from transformers import pipeline
classifier=pipeline("text-classification",model="mrm8488/bert-tiny-finetuned-sms-spam-detection")
msg=["sun rises in the morning","Hey ,are we still meeting for lunch today?","Urgent your account has been suspended"]
for m in msg:
  result=classifier(m)
  print(result)
  print(result[0]["label"])
  print(result[0]["score"])
classifier1=pipeline("text-classification")
result=classifier1("I hate to draw")
print(result)
```

```python
qa_pipeline=pipeline("question-answering")
context="python was created by Guido van rossum in 1991."
context1="i am suriya priya.i live in madurai"
question="Who created python?"
question1="what was created in 1991"
question2="who lives in madurai"
print(qa_pipeline(question=question,context=context))
print(qa_pipeline(question=question2,context=context1))
print(qa_pipeline(question=question1,context=context))
```

```python
generator=pipeline("text-generation",model="gpt2")
text="There are times when the night sky glows with bands of color. The bands may begin as cloud shapes and then spread into a great arc across the entire sky. They may fall in folds like a curtain drawn across the heavens. The lights usually grow brighter, then suddenly dim. During this time the sky glows with pale yellow, pink, green, violet, blue, and red. These lights are called the Aurora Borealis. Some people call them the Northern Lights. Scientists have been watching them for hundreds of years. They are not quite sure what causes them. In ancient times people were afraid of the Lights. They imagined that they saw fiery dragons in the sky. Some even concluded that the heavens were on fire."
print(generator("I love to draw",max_new_tokens=10))
summarizer=pipeline("summarization",model="facebook/bart-large-cnn")
print(summarizer(text,max_length=50,min_length=25))
```

```python
classifier=pipeline("zero-shot-classification",models="facebook/bart-large-cnn")
sequence="I love to do watch movies and series and to do pencil sketching"
labels=["education","entertainment","sports"]
print(classifier(sequence,labels))
```

```python
translator=pipeline("translation",model="facebook/mbart-large-50-many-to-many-mmt")
text="Artificial intelligence is transforming the world."
translation=translator(text,src_lang="en_XX",tgt_lang="fr_XX")
print("Original:",text)
print("French translation:",translation[0]["translation_text"])
translation=translator(text,src_lang="en_XX",tgt_lang="hi_IN")
print(translation)
```

```python
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
labels = ["sports", "technology", "cooking"]
sentences = [
    "python ai model",
    "cricket is a sport",
    "i love to cook"
]
for sent in sentences:
    result = classifier(sent, labels)
    print(f"Sentence: '{sent}' -> {result['labels'][0]} (score: {result['scores'][0]:.2f})")
```
